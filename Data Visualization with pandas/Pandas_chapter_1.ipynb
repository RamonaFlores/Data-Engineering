```python
# Sort homelessness by individuals
homelessness_ind = homeless_df.sort_values("individuals",ascending=False)

# Print the top few rows
homelessness_ind.head()
```

```python
# Sort homelessness by descending family members
homelessness_fam = homeless_df.sort_values("family_members",ascending=False)
homelessness_fam.head()
```

```python
# Sort homelessness by region, then descending family members
homelessness_reg_fam = homeless_df.sort_values(["region","family_members"],ascending=[True,False])


# Print the top few row
homelessness_reg_fam.head()
```

```python
# Select the individuals column
individuals = homeless_df["individuals"]

print(individuals.head())
```

```python
# Select only the individuals and state columns, in that order
ind_state = homeless_df[["individuals","state"]]

ind_state.head()
```

```python
# Filter for rows where individuals is greater than 10000
ind_gt_10k = homeless_df[homeless_df["individuals"]>10000]

# See the result
ind_gt_10k
```

```python
# Filter for rows where region is Mountain
mountain_reg = homeless_df[homeless_df["region"]=="Mountain"]

# See the result
mountain_reg.head()
```

```python
# Filter for rows where family_members is less than 1000
# and region is Pacific
fam_lt_1k_pac = homeless_df[(homeless_df["family_members"]<1000) & (homeless_df["region"]=="Pacific")]

# See the result
fam_lt_1k_pac
```

```python
# The Mojave Desert states
canu = ["California", "Arizona", "Nevada", "Utah"]

# Filter for rows in the Mojave Desert states
mojave_homelessness = homeless_df[homeless_df["state"].isin(canu)]

# See the result
mojave_homelessness
```

##Adding new columns

```python
# Add total col as sum of individuals and family_members
homeless_df["total"]=homeless_df["family_members"]+homeless_df["individuals"]

# Add p_homeless col as proportion of total homeless population to the state population
homeless_df["p_homeless"]=homeless_df["total"]/homeless_df["state_pop"]

# See the result
homeless_df
```

##Combo-attack!

```python
homelessness=homeless_df
# Create indiv_per_10k col as homeless individuals per 10k state pop
homelessness["indiv_per_10k"] = 10000 * homelessness["individuals"] / homelessness["state_pop"]

# Subset rows for indiv_per_10k greater than 20
high_homelessness = homelessness[homelessness["indiv_per_10k"]>20]

# Sort high_homelessness by descending indiv_per_10k
high_homelessness_srt = high_homelessness.sort_values("indiv_per_10k",ascending=False)

# From high_homelessness_srt, select the state and indiv_per_10k cols
result = high_homelessness_srt[["state","indiv_per_10k"]]

# See the result
result
```
