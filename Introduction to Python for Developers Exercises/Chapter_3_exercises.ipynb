```python
import pandas as pd
import matplotlib.pyplot as plt
import sys

def analizar_gravedad(archivo_csv):
    # Cargar los datos
    df = pd.read_csv(archivo_csv)

    # Verificar que contenga la columna esperada
    if 'Absolute acceleration (m/s^2)' not in df.columns:
        raise ValueError("El archivo no contiene la columna 'Absolute acceleration (m/s^2)'")

    # Calcular estadsticas descriptivas
    estadisticas = df['Absolute acceleration (m/s^2)'].describe()
    print("Estadsticas descriptivas de la aceleracin absoluta:")
    print(estadisticas)

    # Calcular medidas adicionales
    moda = df['Absolute acceleration (m/s^2)'].mode().values
    print(f"\nModa: {moda}")
    print(f"Varianza: {df['Absolute acceleration (m/s^2)'].var()}")
    print(f"Desviacin estndar: {df['Absolute acceleration (m/s^2)'].std()}")

    # Crear grfico de barras
    plt.figure(figsize=(10, 6))
    df['Absolute acceleration (m/s^2)'].plot(kind='hist', bins=30, edgecolor='black')
    plt.title('Distribucin de la aceleracin absoluta (gravedad)')
    plt.xlabel('Aceleracin absoluta (m/s)')
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("grafico_gravedad.png")
    print("\nGrfico guardado como 'grafico_gravedad.png'")

# Si se ejecuta desde la lnea de comandos con un argumento
if __name__ == "__main__":
    if len(sys.argv) < 2:
        print("Uso: python script.py archivo.csv")
    else:
        archivo_csv = sys.argv[1]
        analizar_gravedad("acc.csv")
```
