##Scheduling

* Can apply to any task listed in data processing

* Scheduling is the glue of your system

* Holds each piece and organize how they work together

* Runs tasks in a specific order and resolves all dependencies

###Manual, time and sensor scheduling

**Manually**:
* Humans can manually update things, but we'd like to automatize as most of the process as posible.

Sensor Scheduling:
* Set some tasks to execute if a specific condition is met.

For example, we could update the departments table only if a new employee was added to the employees table. No reason to update otherwise

Manual and automated systems can work together!

If a user manually upgrades their membership, then automatically the system has to propagate this information to other parts of the system, to unlock new features and update billing information

###Batches and streams


Data can be sent in:
* Batches:
  * Group records at intervals
  * Often cheaper
  * Some examples:
    * Songs uploaded by artists

    * Info on the employee table

    * Info on the Revenue table

* Streams:

  * Send individual records right away


*Now on batches vs streams.*

If a spotify user wants to listen to a song online we'll send parts of the song to them, but if a user is listenning online then we'll have to batch all parts of the song together so they can save it!

##Parallel Computing

* Basis of modern data processing tools

* Necessary:
  * Mainly because of memory
  * Also for processing power

* How it works:
  * Split tasks up several smaller subtasks

  * Distribute these subtasks over several computers

## Cloud Computing for data processing

Servers on premises:
* Expensive

* Needs space

* Has an electrical and maintenance cost

* Needs enough power for peak moments

* Processing power wasted at quiter times

Servers on the cloud:

* Rented

* Don't need space

* Use just the resources we need and when we need them

* The closer the user the better
