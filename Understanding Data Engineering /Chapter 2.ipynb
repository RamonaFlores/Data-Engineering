##Structured Data

This type of **Data** is entered following a rigid structure, like a spreadsheeet where there are set columns:

* It's easy to search and organize

* It has a consistent model, rows and columns.

* It has defined types

* It's stored in relational databases

* About 20% of the data is structured

* It's created and queried using SQL (*Structured Query Language*)

##Unstructured Data



*  Does not follow a model, can't be contained in row and columns

* Difficult to search and organize

* Usually text, sound , pictures or videos

* Usually stored in data lakes, can appear in data warehouses or databases

* Most data is unstructured

* Can be extremely valuable

###How to add structure?


* Use AI to search and organize unstructured data

* Add information to make it semi-structured

##SQL
* Structured query Language

* Industry standard for Relational Database Management System (RDBMS)

* Allows you to access many records at once, and group, filter or aggregate them.


* Close to written English, easy to write and understand

* Used by **Data engineers** to create and maintain databases




###SQL for data Engineers



* Data engineers use SQL to create, maintain and update tables


```python
CREATE TABLE employees (
    employee_id INT,
    first_name VARCHAR(255),
    last_name VARCHAR(255),
    email VARCHAR(255),
    hire_date DATE,
    department_id INT,
    full_time BOOLEAN,
    office VARCHAR(255)
);
```

###SQL for data scientists



* Data scientist use SQL to query, fitler, group and aggregate data in tables

```python
SELECT first_name, last_name
FROM employees
WHERE role LIKE '%Data%'
```

### A data catalog must exist within a data lake

* What is the source of this data?

* Where is this data used?

* Who is the owner of the data?

* How often is this data updated?

* Good practice in terms of data governance

* Ensures reproducibility

* No catalog -> Risk of turning into a data swamp
